```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,20,100)

plt.figure(dpi=80)
plt.plot(x,x)
plt.title("Función Lineal ${f(x)=x}$");
plt.ylabel("Y") 
plt.xlabel("X")
plt.xticks([0,5,20]);
```

```python
def cudratica(x):
    return x**2

x = np.linspace(-20,20,15)
y = cudratica(x)

plt.plot(x,y, color="#788AA3", linewidth=2,marker='o', linestyle='dashed' )
plt.title("Función Cudrática")
```

```python
plt.figure(figsize=(10,8))

# Gráfico de Puntos
x2= [-10, 10]
y2= [300,300]
plt.scatter(x2,y2, color="#B2C9AB", s=200 )
plt.scatter(x2,y2, color="black", s=100 )
plt.scatter(x2,y2, color="red", s=20 )


# Funciones Cuadraticas
x = np.linspace(-20,20,20)
y = cudratica(x)
yabajo =  cudratica(x)-30
plt.plot(x,y, color="black", linewidth=3,marker='o', linestyle='dashed' )
plt.plot(x,yabajo, color="black", linewidth=3,marker='o', linestyle='dashed' )

plt.title("Carita Traviesa")
plt.savefig("carita.png")
```

## Gráficos estadísticos

```python
chis = np.random.chisquare(df=2, size=(1000,1))

plt.hist(chis, bins=30, density=True, color="green");
```

```python
plt.hist(chis, bins=30, density=True, color="green", cumulative=True);
```

```python
plt.figure(dpi=100)
p = np.random.uniform(size=1000)
n = np.random.poisson(lam=3 ,size=1000)
binomial = np.random.binomial(n, p, size=1000)

plt.hist(binomial, bins=30, density=True, color="#E8DDB5");
plt.xticks(np.arange(0,9));
plt.title("Montecarlo")
plt.xlabel("binomaial~X")
```
